```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Load datasets
sales_data = pd.read_csv('sales_data.csv')  # Load sales dataset
ingredients_data = pd.read_csv('Pizza_ingredients - Pizza_ingredients.csv')  # Load ingredients dataset

# Check for missing values
print("Sales Data Missing Values:\n", sales_data.isnull().sum())
print("Ingredients Data Missing Values:\n", ingredients_data.isnull().sum())

# Handle missing values by dropping rows with missing data
sales_data = sales_data.dropna()
ingredients_data = ingredients_data.dropna()

# Convert 'order_date' column to datetime format
sales_data['order_date'] = pd.to_datetime(sales_data['order_date'], errors='coerce')

# Extract useful features from Date (Year, Month, Day of the week)
sales_data['Year'] = sales_data['order_date'].dt.year
sales_data['Month'] = sales_data['order_date'].dt.month
sales_data['Day_of_Week'] = sales_data['order_date'].dt.dayofweek

# Merge sales and ingredients datasets on 'pizza_name_id'
merged_data = pd.merge(sales_data, ingredients_data, on='pizza_name_id', how='inner')

# Check data types and correct any inconsistencies
print("Merged Data Types:\n", merged_data.dtypes)
print("First few rows of merged data:\n", merged_data.head()
```

**Outlier Detection**
Outliers can heavily influence predictions. Identifying outliers in quantity and total_price helps clean the data.

```python
# Box plot to detect outliers in quantity
plt.figure(figsize=(8, 6))
sns.boxplot(x='quantity', data=sales_data)
plt.title('Boxplot of Quantity Sold')
plt.show()

# Box plot to detect outliers in total price
plt.figure(figsize=(8, 6))
sns.boxplot(x='total_price', data=sales_data)
plt.title('Boxplot of Total Price')
plt.show()
```

```python
from scipy import stats

# Function to detect outliers using Z-score method
def detect_outliers_z(data, column, threshold=3):
    z_scores = stats.zscore(data[column])
    return data[abs(z_scores) > threshold]

# Detect outliers for 'quantity' and 'total_price'
outliers_quantity_z = detect_outliers_z(sales_data, 'quantity')
outliers_total_price_z = detect_outliers_z(sales_data, 'total_price')

# Display outliers
print("Outliers in Quantity using Z-score method:")
print(outliers_quantity_z)

print("\nOutliers in Total Price using Z-score method:")
print(outliers_total_price_z)
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load your data
sales_data = pd.read_csv('sales_merge.csv')

# Z-score method for outlier detection
z_scores = np.abs(stats.zscore(sales_data['quantity']))
outliers_quantity = sales_data['quantity'][z_scores > 3]

# Remove outliers from quantity
cleaned_data = sales_data[~sales_data['quantity'].isin(outliers_quantity)]

# Cap outliers in total_price
cap_value = cleaned_data['total_price'].quantile(0.95)
cleaned_data['total_price'] = np.where(cleaned_data['total_price'] > cap_value, cap_value, cleaned_data['total_price'])

# Impute outliers in quantity with median
median_quantity = cleaned_data['quantity'].median()
cleaned_data['quantity'] = np.where(cleaned_data['quantity'].isin(outliers_quantity), median_quantity, cleaned_data['quantity'])

# Save the cleaned data to a new CSV file
cleaned_data.to_csv('sales_merge.csv', index=False)

print("Cleaned data saved to 'sales_merge.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error

# Step 1: Load and Prepare Data
# Load historical sales data
sales_data_path = 'sales_merge.csv'  # Update this path
sales_data = pd.read_csv(sales_data_path)

# Convert the 'order_date' column to datetime
sales_data['order_date'] = pd.to_datetime(sales_data['order_date'])

# Set 'order_date' as the index
sales_data.set_index('order_date', inplace=True)

# Aggregating the sales data by date (sum the total sales for each day)
# Assuming 'quantity' represents sales
daily_sales = sales_data.resample('D').sum()['quantity']

# Step 2: Data Preprocessing
# Split data into train and test sets
train_size = int(len(daily_sales) * 0.8)
train, test = daily_sales[:train_size], daily_sales[train_size:]

# Step 3: Sales Prediction with ARIMA
# Fit ARIMA model (adjust parameters as necessary)
arima_model = ARIMA(train, order=(5, 1, 0))  # (p, d, q)
arima_model_fit = arima_model.fit()

# Make predictions for the test set
arima_forecast = arima_model_fit.forecast(steps=len(test))
test = test.to_frame()  # Convert to DataFrame for easier handling
test['ARIMA_Predictions'] = arima_forecast

# Step 4: Sales Prediction with Facebook Prophet
# Prepare data for Prophet
prophet_data = train.reset_index().rename(columns={'order_date': 'ds', 'quantity': 'y'})  # Prophet requires 'ds' and 'y' columns

# Fit the Prophet model
prophet_model = Prophet()
prophet_model.fit(prophet_data)

# Make future dataframe for predictions
future = prophet_model.make_future_dataframe(periods=len(test))
forecast_prophet = prophet_model.predict(future)

# Extract predictions for the test set
test['Prophet_Predictions'] = forecast_prophet['yhat'].iloc[-len(test):].values

# Step 5: Model Evaluation
# Calculate MAPE for ARIMA
mape_arima = mean_absolute_percentage_error(test['quantity'], test['ARIMA_Predictions']) * 100
print(f'MAPE for ARIMA: {mape_arima:.2f}%')

# Calculate MAPE for Prophet
mape_prophet = mean_absolute_percentage_error(test['quantity'], test['Prophet_Predictions']) * 100
print(f'MAPE for Prophet: {mape_prophet:.2f}%')

# Plot Actual vs Predicted for both models
plt.figure(figsize=(12, 6))
plt.plot(test.index, test['quantity'], label='Actual Sales', color='blue')
plt.plot(test.index, test['ARIMA_Predictions'], label='ARIMA Predictions', color='orange')
plt.plot(test.index, test['Prophet_Predictions'], label='Prophet Predictions', color='green')
plt.title('Actual vs Predicted Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python

# Step 6: Finalize the Best Model
if mape_arima < mape_prophet:
    best_model = 'ARIMA'
    best_mape = mape_arima
else:
    best_model = 'Prophet'
    best_mape = mape_prophet

print(f'Best Model: {best_model} with MAPE: {best_mape:.2f}%')

```
